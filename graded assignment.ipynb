Question 2: Use Webscraping to Extract Tesla Revenue Data

Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initializing URL for getting the data...
tesla_data_url = 'https://finance.yahoo.com/quote/TSLA/history'

response = requests.get(tesla_data_url)
soup = BeautifulSoup(response.text,'html.parser')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
df
```

Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

```python
import yfinance as yf
import matplotlib.pyplot as plt
ticker = yf.Ticker('TSLA')
df = ticker.history(period="max")
dataframe = df.reset_index()

# make_graph Custom function as..
def make_graph(df_column_Date, df_column_Close):
    plt.plot(df_column_Date, df_column_Close)
    plt.title('Tesla Stock Closing Prices over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price in USD')
    plt.show()
# Calling the function as..
make_graph(dataframe['Date'],dataframe['Close'])

```
